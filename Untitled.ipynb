```python
def count_points(list_points, M):
    
    counts = np.zeros(M,np.int)
    for mm in range(M):
        counts[mm] = sum([mm in sublist for sublist in list_points])
    return counts

def remove_used_terms(list_points, sublist):
    solution = []
    for point in list_points:
        is_there = sum([elem in point for elem in sublist])
        if not is_there:
            solution.append(point)
    return solution
```

```python
#Recursive function
def test(list_points, solution, global_solution, M, counter):
#     print('Function ' + str(counter))
    points = list(range(M))
    counts = count_points(list_points, M)
#     print(counts)
    #Return condition
    if not sum(counts):
#         print('End level ' + str(counter))
        return True
    
    #Get points with less counts, discarding counts=0.
    counts[counts==0] = 1000 #Change this
    ind = np.argmin(counts[counts!=0])
#     print(ind)
    next_point = points[ind]
#     print(next_point)
    #if equal to 1, get the sublist
    if counts[ind] == 1:
        pos = [next_point in sublist for sublist in list_points].index(True)
        sublist = list_points[pos]
        new_list_points = remove_used_terms(list_points, sublist)
#         print('Unique at level ' + str(counter))
        finished = test(new_list_points, solution + [sublist], global_solution, M, counter+1)
        if finished:
            global_solution.append(solution + [sublist])
#             print(global_solution)
#             print('Obtained at level ' + str(counter) + ' from Unique')
            
    else:
        #Get candidate sublists
        candidates = [sublist for sublist in list_points if next_point in sublist]
        for kk, sublist in enumerate(candidates):
            new_list_points = remove_used_terms(list_points, sublist)
#             print('Candidate ' + str(kk) + ' of ' + str(len(candidates)) + ' at level ' + str(counter))
            finished = test(new_list_points, solution + [sublist], global_solution, M, counter+1)
            if finished:
                global_solution.append(solution + [sublist])
#                 print(global_solution)
#                 print('Obtained at level ' + str(counter) + ' from Candidate ' + str(kk))
            
#     print('End level ' + str(counter))
    return False
```
